````python
Q.1. Create two int type variables, apply addition, subtraction, division and multiplications
and store the results in variables. Then print the data in the following format by calling the
variables:
First variable is __ & second variable is __.
Addition: __ + __ = __
Subtraction: __ - __ = __
Multiplication: __ * __ = __
Division: __ / __ = __


ANS:
    Sure! Here's an example of how you can create two `int` variables, perform the arithmetic operations, 
    store the results, and print the data in the requested format:

```python
# Create two int variables
first_variable = 10
second_variable = 5

# Perform arithmetic operations and store the results
addition = first_variable + second_variable
subtraction = first_variable - second_variable
multiplication = first_variable * second_variable
division = first_variable / second_variable

# Print the data in the required format
print(f"First variable is {first_variable} & second variable is {second_variable}.")
print(f"Addition: {first_variable} + {second_variable} = {addition}")
print(f"Subtraction: {first_variable} - {second_variable} = {subtraction}")
print(f"Multiplication: {first_variable} * {second_variable} = {multiplication}")
print(f"Division: {first_variable} / {second_variable} = {division}")
```

Output:
```
First variable is 10 & second variable is 5.
Addition: 10 + 5 = 15
Subtraction: 10 - 5 = 5
Multiplication: 10 * 5 = 50
Division: 10 / 5 = 2.0
```

In the division operation, since both variables are of type `int`, the result will be a float because Python 
performs true division by default. If you want to perform integer division and get an integer result, 
you can use the `//` operator instead of `/`.
````

````python
Q.2. What is the difference between the following operators:
(i) ‘/’ & ‘//’
(ii) ‘**’ & ‘^’


ANS:
    (i) The operators '/' and '//' are both used for division in Python, but they have different behaviors.

- The '/' operator performs normal division and returns the result as a floating-point number. For example,
 if you divide 10 by 3 using '/', the result will be approximately 3.3333333333333335.

- The '//' operator performs floor division or integer division. It divides the left operand by the right operand 
  and returns the largest integer that is less than or equal to the result. For example, if you divide 10 by 
    3 using '//', the result will be 3. It essentially truncates the decimal part of the division result.

Here's an example to illustrate the difference:

```python
x = 10 / 3  # Normal division
print(x)    # Output: 3.3333333333333335

y = 10 // 3  # Floor division
print(y)    # Output: 3
```

(ii) The operators '**' and '^' are both used for exponentiation or raising a number to a power, but they have 
different meanings in Python.

- The '**' operator is the exponentiation operator in Python. It raises the left operand to the power of the right
operand. For example, 2 ** 3 equals 8, as 2 raised to the power of 3 is 8.

- The '^' operator, however, does not represent exponentiation in Python. In Python, the '^' operator is the 
bitwise XOR operator. It performs the bitwise XOR operation on the binary representations of the operands. 
It is used for manipulating individual bits in binary numbers.

Here's an example to illustrate the difference:

```python
x = 2 ** 3  # Exponentiation
print(x)    # Output: 8

y = 2 ^ 3   # Bitwise XOR
print(y)    # Output: 1
```

Note that if you are looking for exponentiation in Python, you should use the '**' operator, not '^'.
````

````python
Q.3. List the logical operators.

ANS:
    The logical operators are used in computer programming to perform logical operations on boolean values 
    (true or false). The most common logical operators are:

1. AND (&&): Returns true if both operands are true, otherwise returns false.
2. OR (||): Returns true if at least one of the operands is true, otherwise returns false.
3. NOT (!): Reverses the boolean value of the operand. If the operand is true, NOT returns false, and if the 
  operand is false, NOT returns true.

Additionally, some programming languages may have additional logical operators such as:

4. XOR (exclusive OR): Returns true if exactly one of the operands is true, otherwise returns false.
5. NAND (not AND): Returns true if at least one of the operands is false, otherwise returns false.
6. NOR (not OR): Returns true if both operands are false, otherwise returns false.

It's worth noting that the availability and syntax of logical operators may vary slightly depending on the 
programming language you are using.
````

````python
Q.4. Explain right shift operator and left shift operator with examples.

ANS:
    The right shift (>>) and left shift (<<) operators are bitwise operators used in computer programming to shift
    the bits of a binary number to the right or left, respectively. These operators operate on the individual bits 
    of an integer or binary representation of a value.

1. Right Shift (>>):
The right shift operator (>>) moves the bits of a number to the right by a specified number of positions. 
It effectively divides the number by 2 raised to the power of the specified shift amount.

Example:
Let's consider the decimal number 12, which is represented in binary as 1100. Performing a right shift by 2 
positions using the right shift operator (12 >> 2) would result in shifting the bits to the right by two positions,
discarding the rightmost bits and introducing zeros on the left side. So, the result would be 0011, which is the 
binary representation of the decimal number 3.

2. Left Shift (<<):
The left shift operator (<<) moves the bits of a number to the left by a specified number of positions. It 
effectively multiplies the number by 2 raised to the power of the specified shift amount.

Example:
Consider the decimal number 5, which is represented in binary as 101. Performing a left shift by 3 positions using
the left shift operator (5 << 3) would result in shifting the bits to the left by three positions and appending
three zeros on the right side. So, the result would be 101000, which is the binary representation of the decimal 
number 40.

In both cases, it's important to note that the shifted bits are discarded or padded with zeros based on the shift
direction and the number of positions specified.

The right shift and left shift operators are commonly used in low-level programming, bitwise operations, and 
optimizing certain algorithms.
````

````python
Q.5. Create a list containing int type data of length 15. Then write a code to check if 10 is
present in the list or not.

ANS:
    Sure! Here's an example code snippet in Python that creates a list of 15 integers and checks if 10 is present
    in the list:

```python
# Create a list of integers
my_list = [5, 3, 8, 2, 10, 7, 12, 15, 20, 17, 6, 13, 9, 4, 1]

# Check if 10 is present in the list
if 10 in my_list:
    print("10 is present in the list.")
else:
    print("10 is not present in the list.")
```

In this code, we create a list called `my_list` that contains 15 integer values. Then we use the `in` operator to
check if 10 is present in the list. If it is, we print a message stating that 10 is present. Otherwise, we print
a message stating that 10 is not present.

You can modify the contents of `my_list` or the number being checked to suit your specific requirements.
````
