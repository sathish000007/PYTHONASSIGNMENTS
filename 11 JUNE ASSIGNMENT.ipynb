````python
1. What is a lambda function in Python, and how does it differ from a regular function?

ANS:
    In Python, a lambda function is a small, anonymous function that can be defined without a name. It is also 
    known as an "anonymous function" because it doesn't require the use of the `def` keyword to define a function
    name. Instead, lambda functions are created using the `lambda` keyword, followed by a list of parameters, 
    a colon (:), and an expression that is executed when the function is called.

Here's the general syntax of a lambda function:

```
lambda arguments: expression
```

Lambda functions are commonly used when you need a simple, one-line function that doesn't need to be defined 
separately. They are typically used in situations where a function is required as an argument to another function 
or when you want to create a function quickly without defining a named function.

Lambda functions are different from regular functions in a few ways:

1. Syntax: Lambda functions have a more concise syntax compared to regular functions. They consist of a single 
expression, whereas regular functions can have multiple statements and a return statement.

2. Namelessness: Lambda functions are anonymous, meaning they don't have a name. They are defined on the spot and 
can only be referred to through the variable they are assigned to or used directly as an argument.

3. Scope: Lambda functions are defined in their own temporary scope and don't have access to variables outside 
their scope unless explicitly passed as arguments.

4. Function Objects: Lambda functions return a function object that can be assigned to a variable, passed as an
argument, or used directly.

Here's an example to illustrate the usage of a lambda function:

```python
# Regular function
def square(x):
    return x ** 2

print(square(5))  # Output: 25

# Lambda function
square_lambda = lambda x: x ** 2

print(square_lambda(5))  # Output: 25
```

In this example, `square()` is a regular function that takes a parameter `x` and returns the square of `x`. The
lambda function `square_lambda` achieves the same functionality but is defined using the lambda syntax. Both 
functions can be called with the same result.
````

````python
2. Can a lambda function in Python have multiple arguments? If yes, how can you define and use
them?
ANS:
    Yes, a lambda function in Python can have multiple arguments. To define a lambda function with multiple 
    arguments, you simply list the arguments separated by commas after the `lambda` keyword. The arguments are
    followed by a colon (:), and then you specify the expression that will be executed when the lambda function 
    is called.

Here's the syntax for a lambda function with multiple arguments:

```python
lambda arg1, arg2, ... : expression
```

You can then use the lambda function by calling it with the appropriate arguments. Here's an example to demonstrate 
this:

```python
# Lambda function with multiple arguments
addition = lambda x, y: x + y

result = addition(3, 5)
print(result)  # Output: 8
```

In this example, the lambda function `addition` takes two arguments `x` and `y` and returns their sum. When the
lambda function is called with arguments `3` and `5`, it returns the result `8`. The same concept can be extended 
to lambda functions with any number of arguments by separating them with commas in the lambda definition.

It's worth noting that lambda functions are typically used for simple and concise operations. If you find yourself
needing more complex functionality or multiple lines of code, it's often better to use a regular named function 
instead of a lambda function.
````

````python
3. How are lambda functions typically used in Python? Provide an example use case.

ANS:
    Lambda functions in Python are typically used for situations where you need a small, anonymous function without 
    the need for a formal function definition. They are commonly employed in functional programming paradigms, where 
    functions are treated as first-class citizens and can be passed as arguments or returned as values.

One example use case for lambda functions is in conjunction with higher-order functions like `map()`, `filter()`, and 
`reduce()`. These functions often expect a function as an argument to perform some operation on elements of a 
sequence. Using lambda functions, you can define the required functionality concisely without the need for a 
separate function definition.

Here's an example to demonstrate the use of lambda functions with `map()`:

```python
numbers = [1, 2, 3, 4, 5]

squared_numbers = list(map(lambda x: x ** 2, numbers))
print(squared_numbers)  # Output: [1, 4, 9, 16, 25]
```

In this example, the lambda function `lambda x: x ** 2` is passed as an argument to `map()`. The lambda function 
squares each element `x` in the `numbers` list. The result is a new list `squared_numbers` that contains the 
squared values of the original numbers.

Lambda functions, in this case, provide a concise way to express the squaring operation without the need to define
a separate named function.

Overall, lambda functions are commonly used in Python to define small, inline functions for one-time or specialized 
operations, particularly when working with higher-order functions or functional programming paradigms.
````

````python
4. What are the advantages and limitations of lambda functions compared to regular functions in
Python?
ANS:
    Lambda functions, also known as anonymous functions, are a concise way of defining small, one-line functions
    in Python. While they offer some advantages over regular functions, they also come with limitations. Here are 
    the advantages and limitations of lambda functions compared to regular functions in Python:

Advantages of Lambda Functions:
1. Concise syntax: Lambda functions are defined in a compact syntax, allowing you to define simple functions in a 
single line of code. This can make your code more readable and reduce the need for defining separate functions for
small tasks.

2. No need for a return statement: Lambda functions automatically return the result of the expression they evaluate.
You don't need to explicitly use a `return` statement as you would in a regular function. This simplifies the code 
and makes it more concise.

3. Functional programming: Lambda functions are often used in conjunction with higher-order functions like `map()`,
`filter()`, and `reduce()`. They enable a functional programming style by allowing you to pass functions as 
arguments to other functions. This can lead to more expressive and flexible code.

4. Local function definition: Lambda functions are defined locally within the scope where they are used. This means 
you can define them directly at the point of use, avoiding the need for a separate function definition elsewhere in
your code.

Limitations of Lambda Functions:
1. Limited functionality: Lambda functions are designed for simple, one-line expressions. They cannot contain
multiple statements, assignments, or complex control flow. If you need more complex logic or multiple statements, 
a regular function is more appropriate.

2. Lack of name: Lambda functions are anonymous, meaning they don't have a name associated with them. This can 
make them harder to debug and trace when an error occurs, as they don't have a meaningful identifier to refer to.

3. Limited readability: While lambda functions can make code more concise, they can also reduce readability, 
especially if the expression becomes too complex or if the lambda function is used in multiple places. In such 
cases, a regular function with a descriptive name might be more readable and maintainable.

4. Reduced reusability: Lambda functions are best suited for short-lived, throwaway functions that are not intended
to be reused. Since they lack a name and are defined locally, it can be challenging to reuse them in different 
parts of your codebase. Regular functions, on the other hand, can be defined globally and can be reused across 
different modules or scripts.

````

````python
5. Are lambda functions in Python able to access variables defined outside of their own scope?
Explain with an example.

ANS:
    Yes, lambda functions in Python can access variables defined outside of their own scope. This is because lambda 
    functions have access to the variables in the enclosing scope, just like regular functions. The concept is known as "lexical scoping" or "closure."

Here's an example to illustrate how lambda functions can access variables from the outer scope:

```python
def outer_function():
    x = 10
    return lambda y: x + y

lambda_func = outer_function()
result = lambda_func(5)
print(result)  # Output: 15
```

In this example, we have a function called `outer_function` that defines a variable `x` and returns a lambda 
function. The lambda function takes an argument `y` and adds it to the variable `x` from the outer scope. 

When `outer_function` is called, it returns the lambda function, which is assigned to the variable `lambda_func`.
Later, we invoke `lambda_func` with an argument of `5`, resulting in `x + y` where `x` is `10` and `y` is `5`. 
Therefore, the output is `15`.

The lambda function has access to the variable `x` even after `outer_function` has completed execution because 
 it retains a reference to the variables in its enclosing scope.

It's important to note that this concept applies to both regular functions and lambda functions. Lambda functions 
are not restricted in their ability to access variables from the outer scope.
````

````python
8. Implement a lambda function to filter out all the even numbers from a list of integers.

ANS:
    Certainly! Here's an example of a lambda function that filters out all the even numbers from a list of integers:

```python
numbers = [10, 5, 8, 15, 3, 12]

even_numbers = lambda lst: list(filter(lambda x: x % 2 == 0, lst))

# Usage example:
result = even_numbers(numbers)
print(result)  # Output: [10, 8, 12]
```

In this example, we have a lambda function called `even_numbers`. It takes a list `lst` as an argument and uses the 
`filter()` function in conjunction with another lambda function to filter out the even numbers.

The inner lambda function `lambda x: x % 2 == 0` checks if each element `x` in the list is divisible by 2 with a 
remainder of 0, i.e., an even number. The `filter()` function then applies this lambda function to each element in
the list and returns a new list containing only the elements that satisfy the condition.

To use the lambda function, we pass a list of integers to it. In this case, `even_numbers(numbers)` filters out 
the even numbers from the `numbers` list, resulting in `[10, 8, 12]`. The filtered list is then printed to the 
console.
````

````python
9. Write a lambda function to sort a list of strings in ascending order based on the length of each
string.
ANS:
    Certainly! Here's an example of a lambda function that sorts a list of strings in ascending order based on the 
    length of each string:

```python
strings = ['apple', 'banana', 'cherry', 'date', 'elderberry']

sorted_strings = lambda lst: sorted(lst, key=lambda x: len(x))

# Usage example:
result = sorted_strings(strings)
print(result)  # Output: ['date', 'apple', 'cherry', 'banana', 'elderberry']
```

In this example, we have a lambda function called `sorted_strings`. It takes a list `lst` as an argument and uses 
the `sorted()` function with a key parameter to specify the sorting criterion.

The key parameter is set to another lambda function `lambda x: len(x)`, which calculates the length of each string
`x`. By using this lambda function as the key, the `sorted()` function will sort the strings in ascending order 
based on their length.

To use the lambda function, we pass a list of strings to it. In this case, `sorted_strings(strings)` sorts the 
`strings` list in ascending order based on the length of each string, resulting in `
['date', 'apple', 'cherry', 'banana', 'elderberry']`. The sorted list is then printed to the console.
````

````python
10. Create a lambda function that takes two lists as input and returns a new list containing the
common elements between the two lists.

ANS:
    Certainly! Here's an example of a lambda function that takes two lists as input and returns a new list
    containing the common elements between the two lists:

```python
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]

common_elements = lambda lst1, lst2: list(filter(lambda x: x in lst2, lst1))

# Usage example:
result = common_elements(list1, list2)
print(result)  # Output: [4, 5]
```

In this example, we have a lambda function called `common_elements`. It takes two lists `lst1` and `lst2` as 
arguments. The lambda function uses the `filter()` function in conjunction with another lambda function to filter 
out the common elements between the two lists.

The inner lambda function `lambda x: x in lst2` checks if each element `x` in `lst1` is present in `lst2`. The 
`filter()` function then applies this lambda function to each element in `lst1` and returns a new list containing 
only the elements that satisfy the condition.

To use the lambda function, we pass two lists to it. In this case, `common_elements(list1, list2)` filters out the 
common elements between `list1` and `list2`, resulting in `[4, 5]`. The filtered list is then printed to the console.
````

````python
11. Write a recursive function to calculate the factorial of a given positive integer.
ANS:
    Certainly! Here's an example of a recursive function to calculate the factorial of a given positive integer in 
    Python:

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

# Usage example:
result = factorial(5)
print(result)  # Output: 120
```

In this example, the `factorial` function is defined recursively. It takes a positive integer `n` as an argument.

The base case of the recursion is when `n` is 0 or 1. In this case, the function simply returns 1 because the 
factorial of 0 or 1 is 1.

For values of `n` greater than 1, the function calls itself with the argument `n - 1` and multiplies the result
by `n`. This recursive step continues until it reaches the base case, and then the calculated factorial value is 
returned.

To use the `factorial` function, we pass a positive integer to it. In this case, `factorial(5)` calculates the
factorial of 5, resulting in 120. The result is then printed to the console.
````

````python
12. Implement a recursive function to compute the nth Fibonacci number.
ANS:
    Certainly! Here's an example of a recursive function to compute the nth Fibonacci number in Python:

```python
def fibonacci(n):
    if n <= 0:
        raise ValueError("n must be a positive integer.")
    elif n == 1 or n == 2:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

# Usage example:
result = fibonacci(6)
print(result)  # Output: 8
```

In this example, the `fibonacci` function is defined recursively. It takes a positive integer `n` as an argument.

The base cases of the recursion are when `n` is 1 or 2. In these cases, the function simply returns 1 because the 
first two Fibonacci numbers are defined as 1.

For values of `n` greater than 2, the function calls itself with the arguments `n - 1` and `n - 2` and returns the 
sum of the results. This recursive step continues until it reaches the base cases, and then the calculated
Fibonacci number is returned.

To use the `fibonacci` function, we pass a positive integer to it. In this case, `fibonacci(6)` computes the 6th Fibonacci number, which is 8. The result is then printed to the console.
````

````python
13. Create a recursive function to find the sum of all the elements in a given list.
ANS:
    Certainly! Here's an example of a recursive function to find the sum of all the elements in a given list in 
    Python:

```python
def sum_list(lst):
    if not lst:
        return 0
    else:
        return lst[0] + sum_list(lst[1:])

# Usage example:
numbers = [1, 2, 3, 4, 5]
result = sum_list(numbers)
print(result)  # Output: 15
```

In this example, the `sum_list` function is defined recursively. It takes a list `lst` as an argument.

The base case of the recursion is when the list is empty. In this case, the function returns 0 as there are no 
elements to sum.

For non-empty lists, the function recursively calls itself with the sublist starting from the second element 
(`lst[1:]`) and adds the first element (`lst[0]`) to the sum of the rest of the list. This recursive step continues
until it reaches the base case, and then the calculated sum is returned.

To use the `sum_list` function, we pass a list to it. In this case, `sum_list(numbers)` computes the sum of all
elements in the `numbers` list, resulting in 15. The sum is then printed to the console.
````

````python
14. Write a recursive function to determine whether a given string is a palindrome.
ANS:
    Certainly! Here's an example of a recursive function to determine whether a given string is a palindrome in 
    Python:

```python
def is_palindrome(string):
    string = string.lower()  # Convert string to lowercase for case-insensitive comparison
    if len(string) <= 1:
        return True
    elif string[0] != string[-1]:
        return False
    else:
        return is_palindrome(string[1:-1])

# Usage example:
result1 = is_palindrome("radar")
print(result1)  # Output: True

result2 = is_palindrome("hello")
print(result2)  # Output: False
```

In this example, the `is_palindrome` function is defined recursively. It takes a string `string` as an argument.

The base case of the recursion is when the length of the string is 0 or 1. In this case, the function returns 
True as a string with 0 or 1 character is considered a palindrome.

For strings with more than 1 character, the function compares the first and last characters of the string. If they 
are not equal, the function returns False as it's not a palindrome. Otherwise, it calls itself with the substring
excluding the first and last characters (`string[1:-1]`) and continues the recursive check.

This recursive step continues until it reaches the base case or encounters a mismatched pair of characters, and 
then it returns the final result.

To use the `is_palindrome` function, we pass a string to it. In the provided usage example, `is_palindrome("radar")`
returns True as "radar" is a palindrome, while `is_palindrome("hello")` returns False as "hello" is not
a palindrome. The results are then printed to the console.
````

````python
15. Implement a recursive function to find the greatest common divisor (GCD) of two positive integers.
ANS:
    Certainly! Here's an example of a recursive function to find the greatest common divisor (GCD) of two positive 
    integers in Python:

```python
def gcd(a, b):
    if b == 0:
        return a
    else:
        return gcd(b, a % b)

# Usage example:
result = gcd(24, 36)
print(result)  # Output: 12
```

In this example, the `gcd` function is defined recursively. It takes two positive integers `a` and `b` as arguments.

The base case of the recursion is when `b` is 0. In this case, the function returns `a` as the GCD between `a`
and 0 is `a`.

For values of `b` greater than 0, the function calls itself with the arguments `b` and the remainder of `a` 
divided by `b` (`a % b`). This recursive step continues until it reaches the base case, and then the GCD is returned.

To use the `gcd` function, we pass two positive integers to it. In this case, `gcd(24, 36)` computes the GCD of 
24 and 36, which is 12. The result is then printed to the console.
````
