````python
Q.5. For what purpose def keyword is used?

ANS:
    The `def` keyword in Python is used to define a user-defined function. It is one of the fundamental 
    keywords in Python for creating reusable blocks of code.

When you use the `def` keyword followed by a function name, parentheses, and a colon, you are declaring a 
function definition. The function definition specifies the name of the function and the block of code that 
will be executed when the function is called.

Here's an example of a simple function definition:

```python
def greet():
    print("Hello, world!")
```

In the above example, `def greet():` declares a function named "greet". The function block is defined by 
the indented code that follows, which in this case is a single line that prints "Hello, world!".

After defining the function, you can call it by using its name followed by parentheses:

```python
greet()
```

When this code is executed, it will call the `greet()` function, which will print "Hello, world!" to the 
console.

The `def` keyword is essential for creating functions in Python, allowing you to encapsulate a set of
instructions into a modular unit that can be called and reused throughout your code. Functions help improve 
code organization, readability, and reusability, making it easier to maintain and manage complex programs.
````

````python
Q.8. Explain the mutable and immutable data types with proper explanation & examples.

ANS:
    In Python, data types can be categorized as either mutable or immutable based on whether their values can 
    be changed after they are created. Understanding the distinction between mutable and immutable data types
    is important because it affects how you work with and modify data in your programs.

1. Immutable Data Types:
   - Immutable data types are those whose values cannot be modified after they are created.
   - Any attempt to change the value of an immutable object results in the creation of a new object with the 
    updated value.
   - Immutable objects are safe from unintended modifications, making them suitable for scenarios where data
    integrity is important.
   - Examples of immutable data types in Python include integers, floating-point numbers, booleans, strings, 
   and tuples.

   Example:
   ```python
   # Example with integers
   x = 5
   print(x)  # Output: 5

   x = x + 1  # Attempt to change the value of x
   print(x)  # Output: 6 (A new object with the updated value is created)
   ```

2. Mutable Data Types:
   - Mutable data types are those that allow their values to be modified after they are created.
   - You can change the value of a mutable object directly without creating a new object.
   - Mutable objects are useful when you need to modify data in-place without creating additional copies.
   - Examples of mutable data types in Python include lists, sets, and dictionaries.

   Example:
   ```python
   # Example with lists
   my_list = [1, 2, 3]
   print(my_list)  # Output: [1, 2, 3]

   my_list.append(4)  # Modifying the list by adding a new element
   print(my_list)  # Output: [1, 2, 3, 4] (The original list is modified)

   my_list[0] = 10  # Modifying an element in the list
   print(my_list)  # Output: [10, 2, 3, 4] (The original list is modified)
   ```

It's important to note that when assigning a new value to a variable that refers to an immutable object, 
a new object is created with the updated value. However, when assigning a new value to a variable that refers
to a mutable object, the original object is modified without creating a new object.

Understanding whether a data type is mutable or immutable is crucial for understanding how variables and 
assignments work in Python, and it impacts how you design your code and handle data manipulation operations.
````

````python
Q.9. Write a code to create the given structure using only for loop.
*
***
*****
*******
*********

ANS:
    Certainly! Here's a code snippet that creates the given structure using only a `for` 
    loop in Python:

```python
num_rows = 5

for i in range(num_rows):
    for j in range(i + 1):
        print('*', end='')
    print()
```

In this code, we use two nested `for` loops. The outer loop iterates over the rows, and the 
inner loop iterates over the columns in each row. The number of columns in each row is 
determined by `i + 1`, where `i` represents the current row number.

The `print('*', end='')` statement prints an asterisk (`*`) without a newline character,
ensuring that the asterisks are printed in the same line. The subsequent `print()` 
statement is used to print a newline character after each row to create the pyramid 
structure.

When you run this code, it will generate the following output:

```
*
**
***
****
*****
```

The number of rows in the pyramid can be adjusted by modifying the `num_rows` variable in
the code snippet.
````

````python
Q.10. Write a code to create the given structure using while loop.
|||||||||
|||||||
|||||
|||
|


ANS:
    
    Certainly! Here's a code snippet that creates the given structure using a `while` loop in Python:

```python
num_rows = 5

i = num_rows
while i >= 1:
    j = i
    while j >= 1:
        print('|', end='')
        j -= 1
    print()
    i -= 1
```

In this code, we use two nested `while` loops. The outer loop iterates over the rows in a descending order starting
from the value of `num_rows`. The inner loop iterates over the columns in each row, printing a pipe symbol
(`|`) for each column.

The `print('|', end='')` statement prints a pipe symbol (`|`) without a newline character. The subsequent 
`print()` statement is used to print a newline character after each row to create the given structure.

When you run this code, it will generate the following output:

```
|||||||||
|||||||
|||||
|||
|
```

The number of rows in the structure can be adjusted by modifying the `num_rows` variable in the code snippet.
````
